````python
import numpy as np

def GCNewton(Hess, Grad, tol):
    z0 = np.zeros_like(Grad)
    r0 = Grad
    d0 = -r0
    iteraciones = 0
    for j in range(len(Grad)):
        dHd = np.dot(d0.T, np.dot(Hess, d0))
        if dHd <= 0:  
            return z0, iteraciones

        alphaj = np.dot(r0.T, r0) / dHd
        z0 = z0 + alphaj * d0
        r1 = r0 + alphaj * np.dot(Hess, d0)
        
        if np.linalg.norm(r1) < tol: 
            return z0, iteraciones
        
        beta = np.dot(r1.T, r1) / np.dot(r0.T, r0)
        d0 = -r1 + beta * d0
        r0 = r1 
        iteraciones += 1

    return z0, iteraciones 

def Backtracking(alpha, rho, c, xk, f, grad, p, N):
    for i in range(N):
        temp = xk + alpha * p
        if f(temp) <= f(xk) + c * alpha * np.dot(grad(xk), p):
            return alpha, True
        alpha *= rho
    return alpha, False

def NewtonTruncado(func, Grad, Hess, x0, tol, N, alpha, rho, c, N_backtracking):
    n = len(x0)
    f0 = func(x0)
    total_iteraciones = 0
    Puntos = []
    for k in range(N):
        Puntos.append(x0)
        gk = Grad(x0)
        if np.linalg.norm(gk) < tol:
            #return x0, gk, Hess(x0), True, total_iteraciones / (k + 1)
            return n, f0, k, func(x0), Puntos[:4] + Puntos[-4:], np.linalg.norm(gk), total_iteraciones/N, True
        
        ek = min(0.5, np.sqrt(np.linalg.norm(gk))) * np.linalg.norm(gk)
        pk, iteraciones_gc = GCNewton(Hess(x0), gk, ek)
        total_iteraciones += iteraciones_gc
        alphak, res = Backtracking(alpha, rho, c, x0, func, Grad, pk, N_backtracking)
        
        if not res:
            #return x0, gk, Hess(x0), False, total_iteraciones / (k + 1)
            return n, f0, k, func(x0), Puntos[:4] + Puntos[-4:], np.linalg.norm(gk), total_iteraciones/N, False
        
        x0 = x0 + alphak * pk
    
    #return x0, gk, Hess(x0), False, total_iteraciones / N
    return n, f0, k, func(x0), Puntos[:4] + Puntos[-4:], np.linalg.norm(gk), total_iteraciones/N, False
#return n, f(x0), k, f(xk), np.linalgnorm(gk), total_iteraciones/N, res
````

````python
#Funciones:

def Cuadratica(x, A, b):
    return 0.5 * np.dot(x.T, np.dot(A, x)) - np.dot(b.T, x)

def GradCuadratica(x, A, b):
    return np.dot(A, x) - b

def HessCuadratica(A):
    return A


def Himmelblau(x):
    return (x[0]**2 + x[1] - 11)**2 + (x[0] + x[1]**2 - 7)**2

def GradHimmelblau(x):
    df_dx1 = 4 * x[0] * (x[0]**2 + x[1] - 11) + 2 * (x[0] + x[1]**2 - 7)
    df_dx2 = 2 * (x[0]**2 + x[1] - 11) + 4 * x[1] * (x[0] + x[1]**2 - 7)
    return np.array([df_dx1, df_dx2])

def HessHimmelblau(x):
    dxx = 12*x[0]**2 + 4*x[1] - 42
    dxy = 4*(x[0]+x[1])
    dyy = 4*x[0] + 12*x[1]**2 - 26
    return np.array([[dxx, dxy], 
                    [dxy, dyy]])



#Funciones para Beale
def Beale(x):
    return ((1.5 - x[0] + x[0]*x[1])**2 +
            (2.25 - x[0] + x[0]*(x[1]**2))**2 +
            (2.625 - x[0] + x[0]*(x[1]**3))**2)

def GradBeale(x):
    a = 2*(1.5 - x[0] + x[0]*x[1])*(-1 + x[1]) + \
        2*(2.25 - x[0] + x[0]*(x[1]**2))*(-1 + (x[1]**2)) + \
        2*(2.625 - x[0] + x[0]*(x[1]**3))*(-1 + (x[1]**3))
    
    b = 2*(1.5 - x[0] + x[0]*x[1])*x[0] + \
        2*(2.25 - x[0] + x[0]*(x[1]**2))*2*x[0]*x[1] + \
        2*(2.625 - x[0] + x[0]*(x[1]**3))*3*x[0]*(x[1]**2)    
    return np.array([a, b])

def HessBeale(x):
    x1, x2 = x[0], x[1]
    H11 = 2 + 2*(1.5 - x1 + x1*x2)**2 + 2*(-1 + x2)**2 + \
          2*(2.25 - x1 + x1*x2**2)**2 + 2*(-1 + x2**2)**2 + \
          2*(2.625 - x1 + x1*x2**3)**2 + 2*(-1 + x2**3)**2
          
    H12 = 2*(1.5 - x1 + x1*x2)*1 + \
          2*(-1 + x2)*x1 + \
          2*(2.25 - x1 + x1*x2**2)*2*x2 + \
          2*(-1 + x2**2)*2*x1*x2 + \
          2*(2.625 - x1 + x1*x2**3)*3*x2**2 + \
          2*(-1 + x2**3)*3*x1*x2**2
    H21 = H12
    
    H22 = 2*(1.5 - x1 + x1*x2)*0 + \
          2*x1*1 + \
          2*(2.25 - x1 + x1*x2**2)*0 + \
          2*2*x1*2*x2 + \
          2*(2.625 - x1 + x1*x2**3)*0 + \
          2*3*x1*(3*x2**2)
          
    return np.array([[H11, H12], [H21, H22]])

#Funcion de Rosenbrock

def Rosenbrock(x):
    x = np.array(x)
    return sum(100*(x[1:] - x[:-1]**2)**2 + (1 - x[:-1])**2)

def GradRosenbrock(x):
    x = np.array(x)
    grad = np.zeros_like(x)
    grad[:-1] = -400 * x[:-1] * (x[1:] - x[:-1]**2) - 2 * (1 - x[:-1])
    grad[1:] += 200 * (x[1:] - x[:-1]**2)
    return grad

def HessRosenbrock(x):
    n = len(x)
    H = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            if i == j:
                H[i, i] = 200
                if i < n - 1:
                    H[i, i] += -400 * (x[i+1] - 3 * x[i]**2)
            elif i + 1 == j or i == j + 1:
                H[i, j] = -400 * x[i] if i < j else -400 * x[j]
    return H
````

````python

N = 5000
e = np.finfo(float).eps
rho = 0.5
c = 0.001
Nb = 500
alpha = 1
#Para función la primera función cuadrática:

#Hacer para n = 10, 100, 1000
for n in [10, 100, 1000]:
    tol = np.sqrt(n)*e**(1/3)
    A = n*np.eye(n) + 1
    b = np.ones(n)
    x0 = np.array([0]*n)


    dimension, f0, it, sol, punt, norm, prom, res = NewtonTruncado(
        lambda x: Cuadratica(x, A, b),
        lambda x: GradCuadratica(x, A, b),
        lambda x: HessCuadratica(A),
        x0, tol, N, alpha, rho, c, Nb
    )

    print(f"Dimension: {dimension}")
    print(f"f(x0): {f0}")
    print(f"Número de iteracioens: {it}")
    print(f"Función evaluada en la solución: {sol}")
    print(f"Primeros y ultimos 4 Xk encontrados: {punt}")
    print(f"Norma del gradiente: {norm}")
    print(f"Promedio de iteraciones por GCNewton: {prom}")
    print(f"¿Convergio?: {res}")
    print("\n")

````

### Solución:

````python
#La función recibe como parámetros la función 𝑓, el punto 𝐱 y el incremento ℎ y devuelve el arreglo de
#tamaño 𝑛 con las aproximaciones de aproximaciones de las derivadas parciales en el punto 𝐱.

def GradienteAn(func, x, h):
    Grad = []
    n = len(x)
    for i in range(n):
        ei = np.zeros(n)
        ei[i] = 1
        Grad.append((func(x + h*ei) - func(x))/ h)
    return Grad
x = np.array([0,0])

#Programar función para Hessiana
#Recibe la función f, punto x, incremento h y devuelve la matriz simétrica de tamaño n
#con las aproximaciones de las segundas derivadas parciales f en el punto x

def HessianaAn(func, x, h):
    n = len(x)
    Hess = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            ei = np.zeros(n)
            ej = np.zeros(n)
            ei[i] = 1
            ej[j] = 1
            Hess[i, j] = (func(x + h*ei + h*ej) - func(x + h*ei) - func(x + h*ej) + func(x))/h**2

    return Hess


#Error relativo = |Valor real - Valor calculado| /Valor real

def errorRelativo_grad(fncf, gradf, n, nt, h):
    '''
    Imprime estadísticas sobre el error relativo entre el gradiente analítico
    de una función y el cálculado con autograd.
    fncf  - Función f
    gradf - Implementación del gradiente analítico de la función f
    n     - Dimensión de la variable n
    nt    - Número de puntos aleatorios en donde se comparan los gradientes 
    '''
    ve = np.zeros(nt)
    print('\nErrores relativos en el cálculo del gradiente:')
    #g_f = GradienteAn(fncf)  # Funcion gradiente generada con autograd
    for it in range(nt):
        x0  = np.random.randn(n)
        g0  = gradf(x0) #Función analítica
        #ga  = g_f(x0)
        ga  = GradienteAn(fncf, x0, h)
        ve[it] = np.linalg.norm(g0-ga)/np.linalg.norm(ga)
    print('Min: %.2e   Media: %.2e    Max: %.2e' %(np.min(ve), np.mean(ve), np.max(ve)))


#Función para error en Hessiana

def errorRelativo_hess(fncf, hessf, n, nt, h):
    '''
    Imprime estadísticas sobre el error relativo entre el gradiente analítico
    de una función y el cálculado con autograd.
    fncf  - Función f
    gradf - Implementación del gradiente analítico de la función f
    n     - Dimensión de la variable n
    nt    - Número de puntos aleatorios en donde se comparan los gradientes 
    '''
    ve = np.zeros(nt)
    print('\nErrores relativos en el cálculo de la Hessiana:')
    #g_f = GradienteAn(fncf)  # Funcion gradiente generada con autograd
    for it in range(nt):
        x0  = np.random.randn(n)
        h0  = hessf(x0) #Función analítica
        #ga  = g_f(x0)
        ha  = HessianaAn(fncf, x0, h)
        ve[it] = np.linalg.norm(h0-ha)/np.linalg.norm(ha)
    print('Min: %.2e   Media: %.2e    Max: %.2e' %(np.min(ve), np.mean(ve), np.max(ve)))

````
